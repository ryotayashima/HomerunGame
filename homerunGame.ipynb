```python
rect = ((0.0, 100.0), (600, 450))
```

```python
def getScreen():
    screen = autopy.bitmap.capture_screen(rect)
    screen = np.array(screen)
    screen = np.reshape(screen, (450,600,3))
    return screen
```

```python
array_to_img(getScreen())
```

```python
def detectImg(img_path, screen):
    template = cv2.imread(img_path, 0)
    img_gray = cv2.cvtColor(screen, cv2.COLOR_BGR2GRAY)
    w, h = template.shape[::-1]
    
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
    threshold = 0.8
    loc = np.where( res >= threshold)
    if len(loc[0]) > 0:
        print('detect '+ str(img_path))
```

```python
screen = getScreen()
detectImg('strike_3.png', screen)
```

```python
time_sta = time.time()
while True:
    screen = getScreen()
    detectImg('foul_6.png', screen)
    detectImg('strike_3.png', screen)
    detectImg('hit_3.png', screen)
    detectImg('homerun_6.png', screen)
    if time.time() - time_sta > 60:
        break
```
